```python
life = pd.read_excel('espc.xlsx')
life
```

```python
datap = life.set_index('country')
datap
```

```python
inicio =  1928
fim    =  2019
passo  =  1

intervalo_anos = np.arange(inicio, fim, passo)

data2 = datap.loc[:, intervalo_anos]
```

```python
paises_nao_membros_new = data2.drop(['Austria', 'Italy', 'Belgium', 'Latvia',
                     'Bulgaria', 'Lithuania', 'Croatia',
                     'Luxembourg', 'Cyprus', 'Malta', 
                      'Netherlands', 'Denmark',
                     'Poland', 'Estonia', 'Portugal', 'Finland',
                     'Romania', 'France', 'Germany',
                     'Slovenia', 'Greece', 'Spain', 'Hungary',
                     'Sweden', 'Ireland', 'United Kingdom'])

paises_membros_new = data2.loc[['Austria', 'Italy', 'Belgium', 'Latvia',
                     'Bulgaria', 'Lithuania', 'Croatia',
                     'Luxembourg', 'Cyprus', 'Malta', 
                      'Netherlands', 'Denmark',
                     'Poland', 'Estonia', 'Portugal', 'Finland',
                     'Romania', 'France', 'Germany',
                     'Slovenia', 'Greece', 'Spain', 'Hungary',
                     'Sweden', 'Ireland', 'United Kingdom'],:]

```

```python
media_paises_nao_membros = paises_nao_membros_new.mean(axis = 0, skipna = True)
media_paises_membrostt     = paises_membros_new.mean(axis = 0, skipna = True)
```

```python
plt.plot(media_paises_membrostt, label = 'Membros', color = 'green')
plt.plot(media_paises_nao_membros, label = 'Não Membros', color = 'blue')
plt.plot([1973,1973],[0,82], 'y--', label = 'Fundação da UE')
plt.plot([1960,1960],[0,82], 'y--', label = 'Pílula Anticoncepcional', color = 'red')
plt.ylim(35,82)
plt.xlabel('Tempo (anos)')
plt.ylabel('Idade Média de Expectativa de Vida')
plt.title('Expectativa de Vida no Mundo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('Tempo (anos)')
ax1.set_ylabel('Idade Média de Expectativa de Vida', color = color)
ax1.plot(intervalo_anos, media_paises_membrostt, color = color)
ax1.tick_params(axis = 'y', labelcolor = color)

ax2 = ax1.twinx()  

color = 'tab:red'
ax2.set_ylabel('Taxa Média de Fertilidade', color = color)  
ax2.plot(intervalo_anos,media_paises_membrosll , color = color)
ax2.tick_params(axis = 'y', labelcolor = color)

fig.tight_layout() 
plt.title('Comparação Taxa de Fertilidade e Expectativa de Vida')
plt.show()
```
